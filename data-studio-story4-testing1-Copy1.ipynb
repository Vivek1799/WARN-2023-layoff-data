```python

def get_table_data(table):
    data = []
    rows = table.find_all('tr')
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        data.append([ele for ele in cols if ele])
    return data
```

```python
import requests
import io
import PyPDF2
import pandas as pd

# Generate a sample DataFrame with PDF links
df = pd.DataFrame({'pdf_links': ['http://www.pdf995.com/samples/pdf.pdf',
                                 'https://www.w3.org/WAI/ER/tests/xhtml/testfiles/resources/pdf/dummy.pdf']})

# Define a function to download and extract text from a PDF file
def extract_text_from_pdf(url):
    response = requests.get(url)
    pdf_file = io.BytesIO(response.content)
    pdf_reader = PyPDF2.PdfFileReader(pdf_file)
    text = ''
    for page_num in range(pdf_reader.getNumPages()):
        page = pdf_reader.getPage(page_num)
        text += page.extractText()
    return text

# Apply the function to the DataFrame and create a new column with the extracted text
df['text'] = df['pdf_links'].apply(extract_text_from_pdf)

# Show the resulting DataFrame
print(df)
```

```python
import os
from tika import parser

# Set the path to the directory where the PDF files are stored
pdf_dir = "all-pdf-files"

# Set the path to the directory where you want to save the text files
txt_dir = "all-text-files"

# Loop through all the PDF files in the directory
for filename in os.listdir(pdf_dir):
    if filename.endswith(".pdf"):
        # Parse the PDF file using Tika
        parsed_pdf = parser.from_file(os.path.join(pdf_dir, filename))
        # Extract the text content from the parsed PDF
        text_content = parsed_pdf['content']
        # Create a new text file with the same name as the PDF file
        txt_filename = os.path.splitext(filename)[0] + ".txt"
        txt_path = os.path.join(txt_dir, txt_filename)
        # Write the text content to the new text file
        with open(txt_path, "w", encoding="utf-8") as f:
            f.write(text_content)
```

```python
import pandas as pd
import re
import glob

# Function to extract required information from the text file
def extract_info(file):
    with open(file, 'r') as f:
        content = f.read()
        
        company_name = re.search(r'Company Name:\s*(.*)', content).group(1)
        business_type = re.search(r'Business Type:\s*(.*)', content).group(1)
        affected_num = re.search(r'Number of Affected:\s*(\d+)', content).group(1)
        layoff_date = re.search(r'Layoff Date:\s*(.*)', content).group(1)

        return {
            'Company Name': company_name,
            'Business Type': business_type,
            'Number of Affected': int(affected_num),
            'Layoff Date': layoff_date
        }

# Read all text files in the directory
txt_files = glob.glob('*.txt')

# Extract the information and store it in a list of dictionaries
data = [extract_info(file) for file in txt_files]

# Create a dataframe from the extracted data
df = pd.DataFrame(data)

# Display the dataframe
print(df)
```

```python
import pandas as pd
import re
import glob

# Function to extract required information from the text file
def extract_info(file):
    with open(file, 'r') as f:
        content = f.read()
        
        def find(pattern):
            result = re.search(pattern, content)
            return result.group(1) if result else None
        
        info = {
            'Date of Notice': find(r'Date of Notice:\s*(.*)'),
            'Amendment': find(r'Amendment:\s*(.*)'),
            'Reason Stated for Filing': find(r'Reason Stated for Filing:\s*(.*)'),
            'Company': find(r'Company:\s*(.*)'),
            'County': find(r'County:\s*(.*)'),
            'Business Type': find(r'Business Type:\s*(.*)'),
            'Number Affected': find(r'Number Affected:\s*(\d+)'),
            'Total Employees': find(r'Total Employees:\s*(\d+)'),
            'Layoff Date': find(r'Layoff Date:\s*(.*)'),
            'Closing Date': find(r'Closing Date:\s*(.*)'),
            'Reason for Dislocation': find(r'Reason for Dislocation:\s*(.*)'),
            'Union': find(r'Union:\s*(.*)'),
            'Classification': find(r'Classification:\s*(.*)')
        }
        
        return info

# Read all text files in the directory
txt_files = glob.glob('all-text-files/*.txt')

# Extract the information and store it in a list of dictionaries
data = [extract_info(file) for file in txt_files]

# Create a dataframe from the extracted data
df = pd.DataFrame(data)

# Display the dataframe
print(df)
```

```python
# create a regex pattern to match the required company names
required_companies_pattern = '|'.join(['Goldman, Sachs & Co., LLC', 'Bed, Bath & Beyond, Inc.', 'Amazon', 'Google'])

# filter the rows where the Company name contains any of the required company names
filtered_df = df[df['Company'].str.contains(required_companies_pattern)]

# print the filtered dataframe
print(filtered_df)
```
